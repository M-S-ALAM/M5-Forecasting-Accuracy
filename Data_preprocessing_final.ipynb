```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os
pd.set_option('display.max_columns', 500)
```

```python
def reduce_mem_usage(df, verbose=False):
    '''
    reduce memory usage by downcasting data types
    from https://www.kaggle.com/harupy/m5-baseline
    '''
    
    start_mem = df.memory_usage().sum() / 1024 ** 2
    int_columns = df.select_dtypes(include=["int"]).columns
    float_columns = df.select_dtypes(include=["float"]).columns

    for col in int_columns:
        df[col] = pd.to_numeric(df[col], downcast="integer")

    for col in float_columns:
        df[col] = pd.to_numeric(df[col], downcast="float")

    end_mem = df.memory_usage().sum() / 1024 ** 2
    if verbose:
        print(
            "Mem. usage decreased to {:5.2f} Mb ({:.1f}% reduction)".format(
                end_mem, 100 * (start_mem - end_mem) / start_mem
            )
        )
    return df
```

```python
%%time
sales=pd.read_csv('sales_train_evaluation.csv').pipe(reduce_mem_usage)
sales.head(2)
```

```python
# To be able to make predictions
# we need to add "test set"
import warnings
warnings.filterwarnings('ignore')
for day in range(1942,1942+28):
    sales['d_' + str(day)] = np.int16(0)
sales.head(2)
```

```python
# We can tranform horizontal representation to vertical "view"
# Our "index" will be 'id','item_id','dept_id','cat_id','store_id','state_id'
# and labels are 'd_' coulmns
def convert(input):
    return input.split('_')[1]
sales=pd.melt(sales,id_vars=['id','item_id','dept_id','cat_id','store_id','state_id'],var_name='d',value_name='demand').dropna()
sales['d'] = sales['d'].apply(convert)
sales['d'] = sales['d'].astype('int16')
sales = sales[sales['d']>1000]# Take day greater than 1000.
```

```python
%%time
calender=pd.read_csv('calendar.csv').pipe(reduce_mem_usage)
calender=calender.fillna('no_event')
calender.head(2)
```

```python
calender['d'] = calender['d'].apply(convert)
calender['d'] = calender['d'].astype('int16')
sales=pd.merge(sales,calender,on='d',how='left').pipe(reduce_mem_usage)
```

```python
%%time
#Load the price datasets.
price=pd.read_csv('sell_prices.csv').pipe(reduce_mem_usage)
price.head(2)
```

```python
sales=pd.merge(sales,price,on=['store_id','item_id','wm_yr_wk'],how='left').pipe(reduce_mem_usage)
sales['sell_price'].fillna(sales.groupby('id')['sell_price'].transform('mean'), inplace=True)
data = sales.copy()
```

```python
data['iteam_sold_avg'] = data.groupby('item_id')['demand'].transform('mean').astype(np.float16)
data['state_sold_avg'] = data.groupby('state_id')['demand'].transform('mean').astype(np.float16)
data['store_sold_avg'] = data.groupby('store_id')['demand'].transform('mean').astype(np.float16)
data['cat_sold_avg'] = data.groupby('cat_id')['demand'].transform('mean').astype(np.float16)
data['dept_sold_avg'] = data.groupby('dept_id')['demand'].transform('mean').astype(np.float16)
data['cat_dept_sold_avg'] = data.groupby(['cat_id','dept_id'])['demand'].transform('mean').astype(np.float16)
data['store_item_sold_avg'] = data.groupby(['store_id','item_id'])['demand'].transform('mean').astype(np.float16)
data['cat_item_sold_avg'] = data.groupby(['cat_id','item_id'])['demand'].transform('mean').astype(np.float16)
data['dept_item_sold_avg'] = data.groupby(['dept_id','item_id'])['demand'].transform('mean').astype(np.float16)
data['state_store_sold_avg'] =data.groupby(['state_id','store_id'])['demand'].transform('mean').astype(np.float16)
data['state_store_cat_sold_avg'] = data.groupby(['state_id','store_id','cat_id'])['demand'].transform('mean').astype(np.float16)
data['store_cat_dept_sold_avg'] = data.groupby(['store_id','cat_id','dept_id'])['demand'].transform('mean').astype(np.float16)
```

```python
%%time
data['rolling_sold_mean'] = data.groupby(['id', 'item_id', 'dept_id', 'cat_id', 'store_id', 'state_id'])['demand'].transform(lambda x: x.rolling(window=7).mean()).astype(np.float16)
```

```python
%%time
from tqdm import tqdm
lags = [1,2,3,4,5,6,7,14,21,28,30,56,63,70]
for lag in tqdm(lags):
    data["lag_" + str(lag)] = data.groupby(['id', 'item_id', 'dept_id', 'cat_id', 'store_id', 'state_id'] ,as_index=False)["demand"].shift(lag).astype(np.float16)
```

```python
%%time
data['expanding_sold_mean'] = data.groupby(['id', 'item_id', 'dept_id', 'cat_id', 'store_id', 'state_id'])['demand'].transform(lambda x: x.expanding(2).mean()).astype(np.float16)
```

```python
%%time
data['daily_avg_sold'] = data.groupby(['id', 'item_id', 'dept_id', 'cat_id', 'store_id', 'state_id','d'])['demand'].transform('mean').astype(np.float16)
```

```python
%%time
data['avg_sold'] = data.groupby(['id', 'item_id', 'dept_id', 'cat_id', 'store_id', 'state_id'])['demand'].transform('mean').astype(np.float16)
data['selling_trend'] = (data['daily_avg_sold'] - data['avg_sold']).astype(np.float16)
```

```python
%%time
means = [5,10,20,28,30,56]
for mns in tqdm(means):
    data['rolling_median_'+str(mns)] = data.groupby(['id', 'item_id', 'dept_id', 'cat_id', 'store_id', 'state_id'])['demand'].transform(lambda x: x.rolling(mns).mean()).astype(np.float16)
```

```python
data_nan = (data.isnull().sum() / len(data)) * 100
data_nan = data_nan.drop(data_nan[data_nan == 0].index).sort_values(ascending=False)
missing_data = pd.DataFrame({'Missing Ratio' :data_nan})
missing_data.head(5)
```

```python
f, ax = plt.subplots(figsize=(10, 5))
plt.xticks(rotation='90')
sns.barplot(x=data_nan.index, y=data_nan)
plt.xlabel('Features ', fontsize=10)
plt.ylabel('Percent of missing features values', fontsize=10)
plt.title('Percent missing data by feature', fontsize=15)
```

```python
for column in missing_data.index:
    print(column)
    data[column] = data[column].astype(float)
    data[column].fillna(data[column].mean(), inplace=True)
```
